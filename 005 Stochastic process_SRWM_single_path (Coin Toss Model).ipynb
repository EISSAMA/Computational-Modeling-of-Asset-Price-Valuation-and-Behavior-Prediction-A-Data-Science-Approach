```python
"""
Simulate Stochastic Process || Symmetric Random Walk Simulation [single path]" 
(Coin Toss Model)
Each toss:
Head (H) → +1
Tail (T) → −1
Random variable:
Sn= sum(i=1-n)Xi
one path
""" 
#load libraries 
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Parameters
n_steps = 10         # number of coin tosses per experiment
```

```python
# Simulate random walks
walks = []

# Generate +1 (Head) or -1 (Tail)
steps = np.random.choice([-1, 1], size=n_steps)
    
# Start at S0 = 0, then cumulative sum
S = np.concatenate(([0], np.cumsum(steps)))
    
walks.append(S)
```

```python
# Time axis from 0 to n_steps
time = np.arange(n_steps + 1)
```

```python
# Plot
plt.figure()
plt.plot(time, S, marker='o')

plt.axhline(0)
plt.title("Single Path Symmetric Random Walk (S₀ = 0)")
plt.xlabel("Step (n)")
plt.ylabel("Sₙ")

plt.grid()
plt.show()
```

```python
# -----------------------------------------
# 📊 Statistics Section # For X_i (steps)
# -----------------------------------------

mean_X = np.mean(steps)
var_X = np.var(steps)


print("Statistics for X_i (coin tosses):")
print(f"Mean = {mean_X:.4f}") 
print(f"Variance = {var_X:.4f}")

```

```python
# ------------------------------------------------------
# 📊 Statistics Section # For S_n (random walk values)
# ------------------------------------------------------

mean_S = np.mean(S)
var_S = np.var(S)

print("\nStatistics for S_n (random walk):")
print(f"Mean = {mean_S:.4f}")
print(f"Variance = {var_S:.4f}")
```

```python
# -----------------------------
# ⚡ Quadratic Variation
# -----------------------------

# increments ΔS_i = S_i - S_{i-1}
increments = np.diff(S)

# quadratic variation
QV = np.sum(increments**2)

print("\nQuadratic Variation:")
print(f"[S]_n = {QV}")
print(f"Theoretical value = {n_steps}")
```
