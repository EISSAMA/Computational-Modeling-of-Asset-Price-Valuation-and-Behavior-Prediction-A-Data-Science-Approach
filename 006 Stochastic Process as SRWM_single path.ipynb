```python
"""
Simulate Stochastic Process || Symmetric Random Walk Simulation" 
(Coin Toss Model)
Each toss:
Head (H) → +1
Tail (T) → −1
Random variable:
Sn= sum(i=1-n)Xi

N path
""" 
#load libraries 
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Parameters
n_steps = 100        # number of coin tosses
n_simulations = 5    # number of random walks
```

```python
# Simulate random walks
walks = []

for _ in range(n_simulations):
    # Generate +1 (Head) or -1 (Tail)
    steps = np.random.choice([-1, 1], size=n_steps)
    
    # Start at S0 = 0, then cumulative sum
    S = np.concatenate(([0], np.cumsum(steps)))
    
    walks.append(S)

walks = np.array(walks)
```

```python
# Time axis from 0 to n_steps
time = np.arange(n_steps + 1)
```

```python
# Plot
plt.figure()

for i in range(n_simulations):
    plt.plot(time, walks[i], label=f'Walk {i+1}')

plt.axhline(0)
plt.title("Symmetric Random Walk (Starting at S₀ = 0)")
plt.xlabel("Step (n)")
plt.ylabel("Sₙ")

plt.legend()
plt.grid()
plt.show()
```
